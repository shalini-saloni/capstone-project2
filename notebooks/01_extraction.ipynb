# Section 5 - Dataset Sourcing

This notebook documents and validates the source dataset used in the capstone ETL pipeline.

```python
# ================================
# 1. Import libraries
# ================================
from pathlib import Path

import pandas as pd
```

```python
# ================================
# 2. Define source path and validate source file
# ================================
source_path = Path('../data/raw/online_retail.csv')

if not source_path.exists():
    raise FileNotFoundError(f'Source file not found: {source_path.resolve()}')

print(f'Source file found: {source_path.resolve()}')
```

```python
# ================================
# 3. Load source dataset
# ================================
# Use ISO-8859-1 encoding for the Online Retail dataset
df_source = pd.read_csv(source_path, encoding='ISO-8859-1')

print(f'Source dataset shape: {df_source.shape}')
```
